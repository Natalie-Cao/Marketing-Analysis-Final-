```python
import pandas as pd 
import datetime
import numpy as np
```

```python
df=pd.read_csv('subscribers.csv')
```

```python
df[:5]
```

```python
df['time']=pd.to_datetime(df['account_creation_date'])
```

```python
df['year']=df['time'].dt.strftime('%Y%m')
```

```python
df['year']
```

```python
Atr_new=df[['year','attribution_survey']].dropna()
```

```python
Atr_new_tec=df[['year','attribution_technical']].dropna()
```

```python
year_list=Atr_new['year'].drop_duplicates()
```

```python
year_list_tech=Atr_new['year'].drop_duplicates()
```

```python
dic1={}
for i in year_list:
    dic1[i]=[]
    
    
```

```python
dic_tech={}
for i in year_list_tech:
    dic_tech[i]=[]
    
```

```python
for i in range(len(Atr_new)):
    try:
        if Atr_new['year'][i] in dic1.keys():
            dic1[Atr_new['year'][i]].append(Atr_new['attribution_survey'][i])
    except:
        pass
```

```python
for i in range(len(Atr_new_tec)):
    try:
        if Atr_new_tec['year'][i] in dic_tech.keys():
            dic_tech[Atr_new_tec['year'][i]].append(Atr_new_tec['attribution_technical'][i])
    except:
        pass
```

```python
for i,v in dic_tech.items():
    method_tech={}
    for n in v:
        method_tech[n]=method_tech.get(n, 0) + 1
        dic_tech[i]=method_tech
    
```

```python
for i,v in dic1.items():
    method={}
    for n in v:
        method[n]=method.get(n, 0) + 1
        dic1[i]=method
```

```python
survey=pd.DataFrame(dic1)
survey.to_csv('survey_number.csv')
```

```python
tech=pd.DataFrame(dic_tech)
tech.to_csv('tech_number.csv')
```

```python
fee_dic={}
for i in date_list:
    fee_dic[i]={}
```

```python
for i in range(40):
        if fee['date'][i] in fee_dic.keys():
            fee_dic[fee['date'][i]].update({fee['channel'][i]:fee['spend_AED'][i]})
```

### Comput the Averge fee by different attribute method

```python
import copy
```

```python
Ave_dic=copy.deepcopy(fee_dic)
```

```python

for k,v in dic1.items():
    for n in v:
        try:
            if fee_dic[k][n]:
                Ave_dic[k][n]=(round(float(fee_dic[k][n]/dic1[k][n]),3))
        except:
            pass
            
```

```python
Ave_dic
```

```python
Ave_dic_tech=copy.deepcopy(fee_dic)
for k,v in dic_tech.items():
    for n in v:
        try:
            if fee_dic[k][n]:
                Ave_dic_tech[k][n]=round(float(fee_dic[k][n]/dic_tech[k][n]),3)
        except:
            pass
```
